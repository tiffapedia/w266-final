```python
import tweets_processor
import importlib
import vocabulary
import utils
import models
import tensorflow as tf
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
import patched_numpy_io
from sklearn.preprocessing import LabelEncoder
from keras.preprocessing.text import Tokenizer
```

```python
importlib.reload(tweets_processor)
#tweets_text, tweets_place = tweets_processor.get_tweets_with_location()
tweets_text, tweets_place = tweets_processor.get_tweets_from_csv('../data/tweets_labelled_balanced.csv')
#print(tweets_text[-1],tweets_place[-1])
```

```python
# create the tokenizer at word level
t = Tokenizer(lower = True, filters ='')
# create the tokenizer at character level
#t = Tokenizer(char_level=True)
t.fit_on_texts(tweets_text)
```

```python
# get the vocab size
vocab_size = len(t.word_counts) + 1
# convert the tweets to sequence of id's
encoded_tweets = t.texts_to_sequences(tweets_text)
```

```python
# split the data
train_data, test_data, train_labels, test_labels = train_test_split(encoded_tweets, tweets_place, test_size=0.33, random_state=0)
```

```python
# padding sequences of smaller
train_x, train_ns = utils.pad_np_array(train_data, max_len=50, pad_id=0)
#print(train_x)
#print(train_tokenized_tweets_ids)
```

```python
# passing through NBOW model
# training
train_labels_array = np.array(train_labels, dtype=np.int32)
model_params = dict(V=vocab_size, embed_dim=50, hidden_dims=[25], num_classes=23,
                    encoder_type='bow',
                    lr=0.1, optimizer='adagrad', beta=0.01)

model = tf.estimator.Estimator(model_fn=models.classifier_model_fn, 
                               params=model_params)

train_params = dict(batch_size=32, total_epochs=20, eval_every=2)

train_input_fn = patched_numpy_io.numpy_input_fn(
                    x={"ids": train_x, "ns": train_ns}, y=train_labels_array,
                    batch_size=train_params['batch_size'], 
                    num_epochs=train_params['eval_every'], shuffle=True, seed=42
                 )

for _ in range(train_params['total_epochs'] // train_params['eval_every']):
    # Train for a few epochs
    model.train(input_fn=train_input_fn)
```

```python
# get test inputs
test_x, test_ns = utils.pad_np_array(test_data, max_len=50, pad_id=0)

# encode test labels
#test_encoded_labels = tweets_processor.encode_labels(test_labels)
```

```python
# evaluate model
test_labels.insert(0, '0')
test_labels_array = np.array(test_labels, dtype=np.int32)
test_input_fn = tf.estimator.inputs.numpy_input_fn(
                    x={"ids": test_x, "ns": test_ns}, y=test_labels_array,
                    batch_size=128, num_epochs=1, shuffle=False
                )

eval_metrics = model.evaluate(input_fn=test_input_fn, name="test")
```
